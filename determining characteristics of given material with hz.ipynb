```python
import math
import cmath



class MatType():                                                                       #Creates data type for materials and their properties
    
    def __init__(self, number, material, er, sigma, meu):
        self.n = number
        self.ma = material
        self.e = er
        self.s = sigma
        self.me = meu


a = MatType(1,  'Water',             80,     5e-4,  1)
b = MatType(2,  'Sea Water',         80,       3,   1)
c = MatType(3,  'Ice',              3.5,    1e-5,   1)
d = MatType(4,  'Clay',              20,       5,   1)
e = MatType(5,  'Barium Titanate', 3279,    1e-6,   1)
f = MatType(6,  'Purified Iron',      1,     1e7, 5e3)
g = MatType(7,  'Mu Metal',           1,   1.6e6, 2e5)
h = MatType(8,  'Copper',             1,  5.96e7,   1)
i = MatType(9,  'Graphite',        12.5,     2e5,   1)
j = MatType(10, 'Diamond',          7.5,   1e-13,   1)
k = MatType(11, 'Glass',             65,   1e-15,   1)
l = MatType(12, 'PTFE',             2.1,   1e-25,   1)


my_object = [a,b,c,d,e,f,g,h,i,j,k,l]                                                        #Creates object array of data type


def input_values():                                                                          #Prints to screen data type for user input

    for obj in my_object:
        print (obj.n, obj.ma)


def input_Check():                                                                           #Checks input to ensure a desired input is chosen

    while (True):
        try:

            matnum = int(input("Enter number of the material you would like to test.\n"))

            if matnum > 12 or matnum < 1:
                while (matnum > 12 or matnum < 1):
                    matnum = int(input("Enter an integer between 1 - 12: "))

            Hz = float(input("Enter the frequency in hertz you would like to analyze the material with. \n"))


            break
        except ValueError:
            print("ENTER AN INT")
        except NameError:
            print("Enter the values as an int.")
        except ZeroDivisionError:
            print("why are you doing math?")
        except SyntaxError:
            print("Dont enter a sentence Doc, enter an int.")
        except TypeError:
            print("Enter an int.")
        except AttributeError:
            print("Enter an integer between 1 - 12: ")

    return matnum, Hz


def eqn_variables(var):                                                                       #Converts user input to desired material type

        if var == 1:
            var = a
        elif var == 2:
            var = b
        elif var == 3:
            var = c
        elif var == 4:
            var = d
        elif var == 5:
            var = e
        elif var == 6:
            var = f
        elif var == 7:
            var = g
        elif var == 8:
            var = h
        elif var == 9:
            var = i
        elif var == 10:
            var = j
        elif var == 11:
            var = k
        elif var == 12:
            var = l

        return var.e, var.s, var.me, var.ma


def math_work(er,sigma, mue, hz):                                                            #Calculates values according to user input


    mueNot = math.pi*4e-7
    mueEq = mue*mueNot

    omega = hz*2*math.pi

    eNot = 8.854e-12
    eEq = eNot*er
    
    loss = sigma/(omega*eEq)
    x = loss

    attin = omega*pow((((mueEq*eEq)/(2))*(pow(1+pow(loss,2),.5) - 1)),.5)
    waveNum =  omega*pow((((mueEq*eEq)/(2))*(pow(1+pow(loss,2),.5) + 1)),.5)
    waveImped = pow(mueEq/eEq, .5)*pow((1-1j*(loss)),-.5)
    propV = omega/waveNum
    waveL = (2*math.pi/waveNum)


    if x == 0:
        x = "a lossles"
    elif x > 100:
        x = "a good conducting"
    elif x < 0.01:
        x = "a low-loss medium"
    else:
        x = "an any medium"


    return attin, waveNum, waveImped, propV, waveL, x


def output(attin, waveNum, waveImped, propV, waveL, matLoss, material, hz):                   #Outputs values to screen for user
    print (material, "is", matLoss, "material at", hz, "hz \n")
    print ("Attenuation constant", attin, "(Np/m)")
    if attin == 0.0:
        print ("Material not loss-less, attenuation approximated to zero")
    print ("\nWave number is:", waveNum, "(rad/m) \n")
    print ("Wave impedance is:", cmath.polar(waveImped), "ohms \n")
    print ("Propagation Velocity:", propV, "(m/s) \n")
    print ("Wavelength is:", waveL, "(m) \n")

def exitprog():
    x = input("Would you like to enter another value? \n")
    
    if x == 'y' or x == 'Y':
        main()

    elif x == 'N' or x == 'n':
        print("Exiting Gracefully")
        exit()
    else:
        print ("Enter y/n")
        exitprog()

def main():

    input_values()                                                                      #Displays input values

    matnum, hz = input_Check()                                                          #Checks input values

    er, sigma, mue, material = eqn_variables(matnum)                                    #Converts input values to correct material type

    attin, waveNum, waveImped, propV, waveL, matLoss = math_work(er,sigma,mue,hz)       #Calculates appropriate mathematical values

    output(attin, waveNum, waveImped, propV, waveL, matLoss, material, hz)              #Outputs calculated values

    exitprog()                                                                          #Reruns or exit program


    

    """for obj in my_object:
        print obj.n"""
    

main()
```
